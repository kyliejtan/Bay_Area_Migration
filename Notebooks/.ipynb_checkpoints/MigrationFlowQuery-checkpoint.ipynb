## Querying the user for the counties they would like to investigate

```python
# Initializing two empty lists to hold the user-defined query states and counties
requested_states_list = []
requested_counties_list = [] 
# Initalizing a variable with the inital condition of the while-loop that will collect all of the user-defined query 
# states and counties
another_request = "y"
# Querying the user for their desired query states and counties using a while-loop
while another_request == "y":
    requested_county = input("What county are you interested in?").title().strip()
    # If the user didn't add "County" after the county name they entered, this conditional statement will add it
    if requested_county.find("County") == -1:
        requested_counties_list.append(requested_county + " County")
    else:
        requested_counties_list.append(requested_county)
    requested_state = input("What state is that county in?").title().strip()
    requested_states_list.append(requested_state)
    another_request = input("Is there another state and county you would like to search for (y/n)?").lower()
    if another_request == "y":
        another_request == "y"
    elif another_request != "y" and another_request != "n":
        print("Please answer y or n")
        another_request = input("Is there another state you would like to search for (y/n)?").lower()
    else:
        break
# Zipping the lists of requested counties and states together
zipped_counties_states = list(zip(requested_counties_list, requested_states_list))
counties_states_dict = {}
# Using a for-loop to add the counties as keys and states as values to the counties_states_dict
for (county, state) in zipped_counties_states:
    counties_states_dict.update({county: state})
# Printing a statement that reads back the dictionary of query counties and states entered by the user
print()
print(f"Here is a dictionary with the counties and states that you entered: {counties_states_dict}") 
#
requested_counties_list.to_csv("Resources/migration_flow_csvs/requested_counties_list.csv", index=False, header=True)
```

## Querying the US Census Bureau ACS Migration Flows API

```python
# Initializing an empty dictionary to hold the query county names and the json responses returned by requests.get()
response_dict = {}
# Using a for-loop to retrieve the state and county (FIPS) numbers from the state and county codes DataFrames
for [county, state] in counties_states_dict.items():
    query_state = state
    query_county = county 
    state_num = state_codes.loc[state_codes["Name"] == query_state, ["State (FIPS)"]].max().item()
    county_num = county_codes.loc[(county_codes["State Code (FIPS)"] == state_num) 
                                                    & (county_codes["Area Name (including legal/statistical area description)"] 
                                                       == query_county), ["County Code (FIPS)"]].max().item() 
    # Using a while-loop to format the state and county nums with the appropriate number of digits
    while len(str(state_num)) < 2:
        state_num_string = "0" + str(state_num)
        state_num = state_num_string
    while len(str(county_num)) < 3:
        county_num_string = "0" + str(county_num)
        county_num = county_num_string
    # Initializing a list that will hold all of the desired query variables
    variable_list = ["COUNTY1_NAME",
                     "COUNTY2_NAME",
                     "STATE1_NAME", 
                     "STATE2_NAME"]
    # Initalizing an empty string that the query variables will join to form the final query variable string
    variables = ""
    # Iterating through the variable_list in order to for the final query variable string
    variables = ""
    for variable in variable_list:
        variables = ",".join(variable_list)
    # Initalizing variables to hold the base url of the ACS Migration Flows API, and the complete query url
    base_url = "https://api.census.gov/data/2015/acs/flows?get="
    query_url = base_url+f"{variables}&for=county:{county_num}&in=state:{state_num}&key={census_key}"
    # Initializing a dictionary that will have the names of our target counties as its keys and the query results as 
    # its values
    # values
    acs_response = requests.get(query_url).json()
    # Updating the response_dict with query counties as keys and the json response from requests.get() as the values
    response_dict.update({query_county: acs_response})
# Exporting the acs query results as a .json file
with open("../Resources/jsons/census_acs_query_result.json", "w") as write_file:
    json.dump(response_dict, write_file)
```

## Optional code to read in the text file containing the merged JSON object so further queries do not have to be made.

```python
# This is an optional piece of code that can be run in cases where the Census Bureau API query has already been 
# carried out and the notebook has been re-opened, so that further blocks of code can be run without having to 
# re-query the API
with open("../Resources/jsons/census_acs_query_result.json", "r") as read_file:
    response_dict = json.load(read_file)
```

## Creating a dictionary of the ACS Migration Flows API responses

```python
# Initializing an empty DataFrame to hold the query counties as keys and the DataFrame versions of the json response
# as values
dataframe_dict = {}
# Using a for-loop to update the dataframe_dict with query counties as keys and the DataFrame versions of the json 
# response as values
for query_county, acs_response in response_dict.items():
    dataframe_dict.update({query_county: pd.DataFrame(acs_response)})
    dataframe_dict[query_county].columns = dataframe_dict[query_county].iloc[0]  
    dataframe_dict[query_county].columns = ["Destination County",
                                            "Origin County",
                                            "Destination State",
                                            "Origin State",
                                            "Destination State Number",
                                            "Destination County Number"]
# Initializing an empty DataFrame to hold the query counties as keys and the series' generated by .groupby as values
dataframe_groupby_dict = {}
# Using a for-loop to update the dataframe_groupby_dict with query counties as keys and the series' generated by 
# .groupby as values
for query_county, dataframe in response_dict.items():
    dataframe_groupby_dict.update({query_county: 
                                   dataframe_dict[query_county].groupby(["Origin State", "Origin County"])\
                                   .count()["Destination County"]})
# Initializing an empty DataFrame to hold the query counties as keys and the descending order-sorted DataFrames as 
# values
dataframe_sorted_dict = {}
# Using a for-loop to update the dataframe_sorted_dict with query counties as keys and the descending order-sorted 
# DataFrames as values
for query_county, dataframe in dataframe_groupby_dict.items():
    dataframe_sorted_dict.update({query_county:
                                  dataframe_groupby_dict[query_county]
                                  .sort_values(ascending=False)
                                  .to_frame()
                                  .reset_index()})
    dataframe_sorted_dict[query_county].columns = ["Origin State", "Origin County", "Count"]
```

## Creating a dictionary of number of people to move into each query county grouped by origin county

```python
# Initializing a variable to hold the name of the first user-defined requested_county so that it can be called as the 
# initial DataFrame on which to join all of the other requested_county DataFrames 
first_requested_county = list(counties_states_dict.keys())[0]
# Initializing an initial DataFrame from the dataframe_sorted_dict on which to outer join the other DataFrames from 
# this dict
combined_counties_df = dataframe_sorted_dict[first_requested_county].iloc[0:10].reset_index()
# Using a for-loop to outer-join all of the DataFrames from the dataframe_sorted_dict
for query_county, dataframe in dataframe_sorted_dict.items():
    combined_counties_df = pd.merge(combined_counties_df, 
                                  dataframe_sorted_dict[query_county].iloc[0:10].reset_index(), how="outer")
# Sorting the combined DataFrame by Origin State
combined_counties_df = combined_counties_df.sort_values("Count", ascending=False)
# Keeping the first instance of each requested_county and droping all other duplicates from the COUNTY2_NAME subset 
# of combined_counties_df 
combined_counties_df = (combined_counties_df.drop_duplicates(subset="Origin County",
                                                        keep='first')
                      .reset_index()
                      .drop(["index", "level_0"], axis=1))
# Renaming the columns of combined_counties_df
combined_counties_df.columns = ["Origin State", "Origin County", "Count"]
```

## Exporting .csv's to be used for statistical analysis and by other notebooks

```python
# Initializing a DataFrame with the sums of people to move from each origin state grouped by those states
counts_by_state_df = combined_counties_df.groupby("Origin State").sum()["Count"]
# Sorting the counts_by_state_df in descending order
counts_by_state_df = counts_by_state_df.sort_values(ascending=False).reset_index()
# Export combined_county_df and counts_by_state_df as CSV's, so they can be read in by another program
combined_counties_df.to_csv("../Resources/migration_flow_csvs/combined_county_df.csv", index=False, header=True)
counts_by_state_df.to_csv("../Resources/migration_flow_csvs/counts_by_state_df.csv", index=False, header=True)
# Initializing an empty DataFrame to hold the origin states as values and the query counties as keys
grouped_by_state_dict = {}
# Using a for-loop to set origin states as values and the query counties as keys
for query_county, dataframe in dataframe_sorted_dict.items():
    # Initializing a variable to hold the reformatted query_county name
    county_name = query_county.replace(" ", "_")
    # Creating a csv for each of the user-defined query_counties
    grouped_by_county_df = pd.DataFrame(dataframe_sorted_dict[query_county])
    grouped_by_county_df.to_csv(f"../Resources/migration_flow_csvs/{county_name}_origin_by_county.csv",
                               index=False, header=True)
    for origin_state in counts_by_state_df:
        grouped_by_state_dict.update({query_county: dataframe_sorted_dict[query_county]
                                                    .groupby("Origin State")
                                                    .max()["Count"]
                                                    .sort_values(ascending=False)
                                                    .reset_index()})
    # Initializing a variable to hold the reformatted query_county name
    county_name = query_county.replace(" ", "_")
    grouped_by_state_df = (dataframe_sorted_dict[query_county].groupby("Origin State")
                                                              .max()["Count"]
                                                              .sort_values(ascending=False)
                                                              .reset_index())
    # Creating a csv for each of the user-defined query_counties
    grouped_by_state_df.to_csv(f"../Resources/migration_flow_csvs/{county_name}_origin_by_state.csv",
                               index=False, header=True)
# Using a for-loop to set origin states as values and the query counties as keys
for query_county, dataframe in dataframe_sorted_dict.items():
    # Initializing a variable to hold the reformatted query_county name
    county_name = query_county.replace(" ", "_")
    grouped_by_county_df = dataframe_sorted_dict[query_county]
    # Creating a csv for each of the user-defined query_counties
    grouped_by_county_df.to_csv(f"../Resources/migration_flow_csvs/{county_name}_origin_by_county.csv",
                                index=False, header=True)
```

## Creating plots that depict the number of people moving into each query county by origin county

```python
# Using a for-loop to plot the number of people who moved to each target county from other states

for query_county in dataframe_sorted_dict:
    ax = plt.subplot()
    # Initializing variables to hold the x and y values to be plotted  #
    # x = dataframe_sorted_dict[query_county][["Origin County", "Origin State"]].iloc[0:20]
    x = dataframe_sorted_dict[query_county]["Origin County"].iloc[0:20]
    y = dataframe_sorted_dict[query_county]["Count"].iloc[0:20]
    query_state = dataframe_sorted_dict[query_county]["Origin State"]
    # Setting the characteristics of each plot
    plt.bar(x, y)
    plt.title(f"Number of People Who Moved to {query_county}")
    plt.ylabel("Number of People")
    plt.xticks(rotation=45, ha="right")
    label_df = dataframe_sorted_dict[query_county][["Origin County", "Origin State"]].iloc[0:20]
    labels = label_df["Origin County"] + ", " + label_df["Origin State"]
    ax.set_xticklabels(labels)
    # Saving each plot as its own image
    formatted_county_name = query_county.replace(" ", "_")
    plt.savefig(f"../Images/FromCountiesBarCharts/to_{formatted_county_name}_from_counties.png",
                dpi=300, bbox_inches = "tight")
    # Displaying each plot at the end of each loop
    plt.show() 
```
